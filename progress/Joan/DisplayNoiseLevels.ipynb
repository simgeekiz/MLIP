```python
from toolkit.file_operations import *
from toolkit.noise import add_noisy_pixels_img, display_digit
import numpy as np
import matplotlib.pyplot as plt
```

```python
[train, _, _] = mnist_to_nparray('data/', 'train', 'test', False)
img = train[9]
```

```python
imgdict = {}
for i in range(20):
    copy = np.copy(img)
    add_noisy_pixels_img(copy, i*.05)
    imgdict['img{0}'.format(i)] = copy
```

```python
for i in range(20):
    plt.subplot(4, 5, i+1)
    digit = imgdict['img{0}'.format(i)].reshape((28,28))
    plt.imshow(digit, cmap='gray')
```
